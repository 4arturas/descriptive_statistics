```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

data = pd.read_csv("D:\PycharmProjects\DescriptiveStatistics\datasets\weight-height.csv")
data.head()
```

```python
data.describe()
```

```python
plt.figure(figsize=(12,8))
sb.countplot( data["Gender"] )
plt.show()
```

```python
plt.figure(figsize=(12,9))
data["Height"].plot(kind="hist", title="Height Histogram")
```

```python
plt.figure(figsize=(12,8))
data["Height"].plot(kind="box", title="Height Box-plot")
```

```python
data["Height"].plot( kind="kde", title="Height KDE", figsize=(12,8) )
```

```python
data["Weight"].plot( kind="kde", title="Weight KDE", figsize=(12,8)  )
```

```python
plt.figure( figsize=(12,8) )
sb.scatterplot( x="Height", y="Weight", data=data )
plt.title("Height vs Weight scatterplot")
```

```python
plt.figure( figsize=(12,8) )
sb.scatterplot( x="Height", y="Weight", hue="Gender", data=data )
# We see from the scatterplot tha heights and weights of males are on average higher than of females, below we 
# and by grouping bellow we see that it is indeed the case
```

```python
gender_groupby = data.groupby("Gender", as_index=False )
gender_groupby.head()
```

```python
gender_groupby.describe().T
```

```python
sb.FacetGrid( data, hue="Gender", height=5 ).map( sb.distplot, "Height").add_legend()
```

```python
sb.FacetGrid( data, hue="Gender", height=5).map( sb.distplot, "Weight").add_legend()
```
