```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

col_gender = "Gender"
col_height = "Height"
col_weight = "Weight"
hwOrig = pd.read_csv('D:\PycharmProjects\DescriptiveStatistics\datasets\Person_Gender_Height_Weight_Index.csv')
 
hw = pd.DataFrame( hwOrig.filter([col_gender,col_height,col_weight], axis=1), columns=[col_gender, col_height, col_weight] )

gender = hwOrig[col_gender] 
height = hwOrig[col_height] 
weight = hwOrig[col_weight]

# hw
```

```python
mean_height = height.mean()
median_height = height.median()
```

```python
plt.figure( figsize=(12,8))
height.hist(bins=30)
plt.axvline( mean_height, color='r', label="mean")
plt.axvline( median_height, color='g', label="median")
plt.title( "Height histogram")
plt.legend()
plt.show()
```

```python
def data_mod( data ) :
    data_counts = {}
    for d in data :
        if ( d not in data_counts ) :
            data_counts[d] = 1
        else :
            data_counts[d] += 1    
    maxKey = -1
    maxValue = -1
    for k in data_counts.keys() :
        v = data_counts[k]
        if v > maxValue :
            maxKey = k
            maxValue = v
    return  maxKey

mod_height = data_mod( height )
mod_height
```

```python
col = col_height
data_for_mod = hw[col]
mod_dict = {}
for d in data_for_mod :
    if d not in mod_dict :
        mod_dict[d] = 1
    else :
        mod_dict[d] += 1
    
max_val = -1
mod_value = -1
idx = 0
mod_idx = 0
for k in mod_dict :
    v = mod_dict[k]
    if v > max_val :
        mod_value = k
        max_val = v
        mod_idx = idx
    idx = idx + 1
    
plt.figure(figsize=(12,8))
x_range = range( len( mod_dict ) )
plt.bar( x_range, list(mod_dict.values()), align="center")
plt.xticks( x_range, list(mod_dict.keys()) )
plt.xlabel( col )
plt.ylabel( "Count" )
plt.axvline( mod_idx, color='red', label="Mod" )
plt.legend()
plt.show()
```
