```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hw = pd.read_csv('D:\PycharmProjects\DescriptiveStatistics\datasets\Person_Gender_Height_Weight_Index.csv')
rows = hw.shape[0]
cols = hw.shape[1]
print( "Rows={0} Cols={1}".format( rows, cols ) )
```

```python
height = hw[["Height"]].copy()
weight = hw[["Weight"]].copy()
```

```python
counts = [1] * rows
height["counts_height"] = counts
weight["counts_weight"] = counts
```

```python
height = height.sort_values("Height")
weight = weight.sort_values("Weight")
```

```python
height = height.groupby("Height", as_index=False).count()
weight = weight.groupby("Weight", as_index=False).count()
```

```python
height["cumcounts_height"] = height["counts_height"].cumsum()
weight["cumcounts_weight"] = weight["counts_weight"].cumsum()
```

```python
q1_height = hw["Height"].quantile(1/4)
q2_height = hw["Height"].quantile(3/4)
q1_weight = hw["Weight"].quantile(1/4)
q2_weight = hw["Weight"].quantile(3/4)
iqr_height = q2_height - q1_height
iqr_weight = q2_weight - q1_weight
```

```python
plt.figure( figsize=(12,8))
hw["Height"].hist(bins=30)
plt.axvline( q1_height, color="r", label="Q1" )
plt.axvline( q2_height, color="g", label="Q2" )
plt.legend()
plt.xlabel("Heights")
plt.ylabel("Counts")
```

```python
plt.figure(figsize=(12,8))
hw["Weight"].hist(bins=30)
plt.axvline( q1_weight, color="m", label="Q1" )
plt.axvline( q2_weight, color="y", label="Q1" )
plt.legend()
plt.xlabel("Weights")
plt.ylabel("Counts")
```

```python
plt.figure( figsize=(12,8) )
plt.scatter( hw["Weight"], hw["Height"], s=100 )
plt.xlabel("Weight")
plt.ylabel("Height")
plt.axvline( q1_weight, color="r", label="Q1" )
plt.axvline( q2_weight, color="g", label="Q2" )
plt.axhline( q1_height, color="m", label="Q1" )
plt.axhline( q2_height, color="y", label="Q2" )
plt.legend()
```

```python
plt.figure(figsize=(12,8))
plt.bar( height["Height"], height["cumcounts_height"] )
plt.xlabel("Height")
plt.ylabel("Cumcounts")
plt.axvline( q1_height, color="r", label="Q1" )
plt.axvline( q2_height, color="g", label="Q2" )
plt.legend()

```
